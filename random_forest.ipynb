```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
y_real = y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_true=y_real, y_pred=y_predicted)
```

```python
import plotly.graph_objs as go
fig = go.Figure(data=[go.Heatmap(z=cm, colorscale='Blues', 
                                 text=cm, texttemplate="%{text}")])
fig.update_layout(title="Confusion Matrix")
fig.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
plotting = ConfusionMatrixDisplay(cm)
plotting.plot()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=digits.target_names)
fig, ax = plt.subplots(figsize=(10,10))
disp.plot(ax=ax, cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
