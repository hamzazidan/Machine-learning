```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
x = np.arange(1, 8)
y = np.array([1.5, 3.8, 6.7, 9, 11.2, 13.6, 16])
print(x, y)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x.reshape(-1, 1), y)
```

```python
%matplotlib inline
plt.grid()
plt.plot(x, y, marker='*', color='red')
```

```python
reg.predict([[3]])
```

```python
reg.coef_, reg.intercept_
```

```python
new_data = np.linspace(1, 10, 1000)
plt.plot(new_data, reg.predict(new_data.reshape(-1, 1)))
plt.scatter(x, y, color='red')
plt.grid()
plt.legend(['fit line', 'real data'])
```
